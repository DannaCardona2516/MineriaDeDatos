## <span style="color:maroon">Introducción a la regresión lineal</span>

La regresión lineal es un método estadístico que trata de modelar la relación entre una variable continua y una o más variables independientes mediante el ajuste de una ecuación lineal. Se llama regresión lineal simple cuando solo hay una variable independiente y regresión lineal múltiple cuando hay más de una. Dependiendo del contexto, a la variable modelada se le conoce como variable dependiente o variable respuesta, y a las variables independientes como regresores, predictores o features. [Vease](https://www.cienciadedatos.net/documentos/py10-regresion-lineal-python.html)

Actualmente se conoce también que la regresión lineal es un algoritmo de [aprendizaje supervisado](https://www.aprendemachinelearning.com/aplicaciones-del-machine-learning/#supervisado) que se utiliza en Machine Learning y en estadística. En su versión más sencilla, lo que haremos es “dibujar una recta” que nos indicará la tendencia de un conjunto de datos continuos (si fueran discretos, utilizaríamos Regresión Logística). Tomado de [Aprende Machine Learning](https://www.aprendemachinelearning.com/regresion-lineal-en-espanol-con-python/).

##   <span style="color:maroon">Un poco de historia</span>

La primera forma de regresión lineal documentada  fue el [método de mínimos cuadrados](https://es.wikipedia.org/wiki/M%C3%ADnimos_cuadrados), que fue publicado por [Legendre](https://es.wikipedia.org/wiki/Adrien-Marie_Legendre) en 1805, y por [Gauss](https://es.wikipedia.org/wiki/Carl_Friedrich_Gauss) en 1809. Legendre y Gauss aplicaron el método para el problema de determinar, a partir de observaciones astronómicas, las órbitas de los cuerpos alrededor del Sol (principalmente cometas, pero también más tarde los entonces recién descubiertos planetas menores). Gauss publicó un desarrollo posterior de la teoría de los mínimos cuadrados en 1821,6​ incluyendo una versión del teorema de Gauss-Markov. [Vease](https://es.wikipedia.org/wiki/Análisis_de_la_regresión#cite_note-Legendre-4)

<center><img src="gauss-legendre.jpg" style="width:400px;heigth:800px"/></center>

El término "regresión" fue acuñado por [Francis Galton](https://es.wikipedia.org/wiki/Francis_Galton) en el siglo xix para describir un fenómeno biológico. El fenómeno fue que las alturas de los descendientes de ancestros altos tienden a regresar hacia abajo, hacia un promedio normal (un fenómeno conocido como [regresión hacia la media](https://es.wikipedia.org/wiki/Regresión_a_la_media) ). Para Galton, la regresión sólo tenía este significado biológico, pero su trabajo fue extendido más tarde por Udny Yule y Karl Pearson a un contexto estadístico más general. En la obra de Yule y Pearson, la distribución conjunta de la variable respuesta y las variables explicativas se supone que es Gaussiana. [Vease](https://es.wikipedia.org/wiki/Análisis_de_la_regresión#cite_note-Legendre-4)

##   <span style="color:maroon">Planteamiento del algoritmo</span>

<span style="color:maroon">Nota:</span> Todo el planteamiento del algoritmo fue tomado de [Regresión lineal con Python](https://www.cienciadedatos.net/documentos/py10-regresion-lineal-python.html) realizado por **Joaquín Amat Rodrigo**.

El modelo de regresión lineal (Legendre, Gauss, Galton y Pearson) considera que, dado un conjunto de observaciones $\{y_i,x_{i1},...,x_{ip}\}_{i=1}^n$, la media $\mu$ de la variable respuesta $y$ se relaciona de forma lineal con la o las variables regresoras $x_1,...,x_p$ acorde a la ecuación:


$$mu_y=\beta_0+\beta_1x_1+\beta_2x_2+\cdot\cdot\cdot+\beta_px_p$$

El resultado de esta ecuación se conoce como la línea de regresión poblacional, y recoge la relación entre los predictores y la **media** de la variable respuesta.

Otra definición que se encuentra con frecuencia en los libros de estadística es:

$$y_i=\beta_0+\beta_1x_{i1}+\beta_2x_{i2}+\cdot\cdot\cdot+\beta_px_{ip}+\epsilon_i$$

En este caso, se está haciendo referencia al valor de $y$ para una observación $i$ concreta. El valor de una observación puntual nunca va a ser exactamente igual al promedio, de ahí que se añada el término de error $\epsilon$.

En ambos casos, la interpretación de los elementos del modelo es la misma:


* $\beta_0$: es la ordenada en el origen, se corresponde con el valor promedio de la variable respuesta $y$ cuando todos los predictores son cero.

* $\beta_j$: es el efecto promedio que tiene sobre la variable respuesta el incremento en una unidad de la variable predictora $x_j$, manteniéndose constantes el resto de variables. Se conocen como coeficientes parciales de regresión.

* $e$: es el residuo o error, la diferencia entre el valor observado y el estimado por el modelo. Recoge el efecto de todas aquellas variables que influyen en $y$ pero que no se incluyen en el modelo como predictores.




En la gran mayoría de casos, los valores  $\beta_0$ y $\beta_j$
poblacionales se desconocen, por lo que, a partir de una muestra, se obtienen sus estimaciones  $\hat{\beta}_0$ y $\hat{\beta}_j$
 . Ajustar el modelo consiste en estimar, a partir de los datos disponibles, los valores de los coeficientes de regresión que maximizan la verosimilitud (likelihood), es decir, los que dan lugar al modelo que con mayor probabilidad puede haber generado los datos observados.
 
 
El método empleado con más frecuencia es el ajuste por mínimos cuadrados ordinarios (OLS), que identifica como mejor modelo la recta (o plano si es regresión múltiple) que minimiza la suma de las desviaciones verticales entre cada dato de entrenamiento y la recta, elevadas al cuadrado. 












<center><img src="modelo.jpg" style="width:500px;heigth:1000px"/></center>

Computacionalmente, estos cálculos son más eficientes si se realizan de forma matricial:

$$y=X^T\beta+\epsilon$$

$$y=\begin{bmatrix}
y_1\\
y_2\\
\cdots\\
y_n
\end{bmatrix}, X=\begin{bmatrix}
1 & x_{11} & \cdots & x_{10}\\
1 & x_{21} & \cdots & x_{2p}\\
\cdots & \cdots & \cdots & \cdots\\
1 & x_{n1} & \cdots & x_{np}
\end{bmatrix}, \beta=\begin{bmatrix}
\beta_0\\
\beta_1\\
\cdots\\
\beta_n
\end{bmatrix}, \epsilon=\begin{bmatrix}
\epsilon_1\\
\epsilon_2\\
\cdots\\
\epsilon_n
\end{bmatrix} $$



$$\hat{\beta}=argmin_{\beta}(y-X^T\beta)^2$$


Una vez estimados los coeficientes, se pueden obtener las estimaciones de cada observación $(\hat{y_i}):$

$$\hat{y}_i=\hat{\beta}_0+\hat{\beta}_1x_{i1}+\hat{\beta}_2x_{i2}+\cdots+\hat{\beta}_px_{ip}$$

Finalmente, la estimación de la varianza del modelo $(\hat{\sigma}^2)$ se obtiene como:

$$\hat{\sigma}^2=\frac{\sum_{i=1}^n\hat{e}_1^2}{n-p}=\frac{\sum_{i=1}^n(y_i-\hat{y}_1)^2}{n-p}$$

donde $n$ es el número de observaciones y $p$
el número de predictores. 



##   <span style="color:maroon">Ejemplo de regresión lineal, con datos reales</span>

Se realiza a continuación un ejemplo de aplicación de una regresión lineal simple con la base de datos **House Prices** tomada de [kaggle](https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data?select=train.csv). Como variable respuesta se tiene el precio de las viviendas notada por **'SalePrice'** y variable regresora la superficie habitable por encima del nivel del suelo en pies cuadrados notada por **'GrLivArea'**. 

### Modelo de regresión

Dadas las variables de estudio se planeta el siguiente modelo de regresión lineal simple

$$
SalePrice_i=\beta_0+\beta_1GrLivArea_i+\epsilon_i
$$

Luego procedemos a hacer todo el análisis respectivo con los datos.

```python
import numpy as np
from numpy.linalg import inv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('train.csv', sep = ',')
data.head()
```

Para darnos una mejor idea de la relación de la variable respuesta y la variable regresora, se realiza un digrama de dispersión en donde se evidencia que hay una relación lineal positiva entre ellas aumentando en el mismo sentido, la pendiente de esta relación es muy acentuada.

```python
plt.figure(figsize = (20, 8))
plt.plot(data.GrLivArea, data.SalePrice, 'co', markersize = 3)
plt.title('Precio de venta vs superficie habitable sobre el nivel del suelo', fontsize = 20)
plt.xlabel('Superficie habitable por encima del nivel del suelo ', fontsize = 20)
plt.ylabel('Precio de venta', fontsize = 20)
```

Dado que los datos atípicos u outliers pueden afectar marcadamente el modelo, además de suponer una fuente de información en sí misma se hace un rapido análisis de estos y luego se eliminan algunos.

Según el diagrama de dispersión anterior se puede notar que:

* Los dos valores más altos de la variable 'GrLivArea' resultan extraños, podría tratarse de terrenos agrícolas o muy degradados, algo que explicaría su bajo precio. Lo que está claro es que estos dos puntos son atípicos, por lo que se  proceder a eliminarlos.


Luego, se realiza un nuevo diagrama de dispersión sin los dos datos antes mencionados.

```python
data = data[~data.Id.isin([1299, 524])]  ### FILTRAR DATOS ATIPICOS
```

```python
plt.figure(figsize = (20, 8))
plt.plot(data.GrLivArea, data.SalePrice, 'co', markersize = 3)
plt.title('Precio de venta vs superficie habitable sobre el nivel del suelo', fontsize = 20)
plt.xlabel('Superficie habitable por encima del nivel del suelo ', fontsize = 20)
plt.ylabel('Precio de venta', fontsize = 20)
```

### Implementación de la Regresion lineal 

Recordando el algoritmo mencionado anteriormente se necesita buscar la estimación de los parametros del modelo.

En este caso el vector $\beta$ esta formado con tan solo dos elementos, esto es:

$$
\beta=\begin{bmatrix}
\beta_0\\
\beta_1
\end{bmatrix}
$$

Siguiendo la estimación de los parametros por el Método de Mínimos Cuadrados, el estimador para el vector $\beta$ esta dado por:

$$
\hat{\beta}=(X^tX)^{-1}X^tY
$$

Por tanto se procede de la siguiente manera

```python
x = data.GrLivArea.to_numpy()
y = data.SalePrice.to_numpy()
```

```python
m = x.shape[0]
```

```python
x = np.array([np.ones(m), x]).T
```

```python
beta = np.dot(inv(np.dot(x.T, x)), np.dot(x.T, y))
beta
```

Se tiene así que las estimaciones de cada observación estan dadas por:

$$
SalePrice_i=7.168,97+115.04GrLivArea_i
$$

Ahora, se muestra graficamente tanto el diagrama de dispersión como la recta de la regresión lineal.

```python
x_teo = np.arange(400, 4500)
y_teo = beta[1]*x_teo + beta[0]
```

```python
plt.figure(figsize = (20, 8))
plt.plot(data.GrLivArea, data.SalePrice, 'co', markersize = 3, label = 'Datos reales')
plt.plot(x_teo, y_teo, 'k', label = 'Ajuste regresion lineal')
plt.title('Precio de venta vs superficie habitable sobre el nivel del suelo', fontsize = 20)
plt.xlabel('Superficie habitable por encima del nivel del suelo ', fontsize = 20)
plt.ylabel('Precio de venta', fontsize = 20)
plt.text(3000, 290000, 'Y= 7168.97*X+ 115.04', fontsize=20)
plt.legend()
```

Finalmente se encuentra el valor del $R^2$ del modelo:

```python
VarX = data.GrLivArea.var()
VarY = data.SalePrice.var()
CovXY = np.cov(data.GrLivArea,data.SalePrice)
CovXY
```

```python
Coef_Determinacion = CovXY[0, 1]**2/(CovXY[0, 0] * CovXY[1,1])
Coef_Determinacion
```

##   <span style="color:maroon">Conclusiones</span>

La regresión lineal simple es la técnica más utilizada, es una forma que permite modelar una relación entre dos conjuntos de variables. El resultado es una ecuación que se puede utilizar para hacer proyecciones o estimaciones sobre los datos.

Como se pudo apreciar en el anterior ejemplo la relación lineal establecida entre el precio de las viviendas y la superficie habitable por encima del nivel del suelo en pies cuadrados sigue la ecuacion de la recta $SalePrice_i=7.168,97+115.04GrLivArea_i$ y su coeficiente de determinación $R^2$ es del 0.54
es decir que el 54% es el porcentaje de la variación en la variable de los precios de vivienda que es explicado por el modelo lineal.

Así se puede concluir que aunque el modelo no es el mejor tampoco se puede asegurar que sea el mejor pues el 54% que se obtuvo con el coeficiente de determinación es bastante regular para los fines pertinenetes que se tengan con el estudio.
